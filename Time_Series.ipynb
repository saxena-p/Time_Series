This notebook contains a series a bunch of codes to work with time series data.

```python
import numpy as np 
import matplotlib.pyplot as plt 
import Sequential_analysis_funcs as sf
```

```python
time = np.arange(0, 100, 0.1)
signal = 3* np.sin(time/10)

sf.plot_series(time, signal, label='my label')
```

```python
# Generate time steps
time = np.arange(4 * 365 + 1)

# Define the parameters of the seasonal data
period = 365
amplitude = 40
slope = 0.05

# Generate the seasonal data
series = sf.trend(time, slope) + sf.seasonality(time, period=period, amplitude=amplitude)

# Plot the results
sf.plot_series(time, series)
```

```python
# Now add noise to the above data.
noise_level = amplitude* 0.02
series += sf.noise(time, noise_level)

# Plot the results
sf.plot_series(time, series)
```

```python
# Now we will use the moving average to remove the noise from the data.
avg_series = sf.moving_average(series, 20)
print("Length of the series is", len(series), "while length of the moving average series is", len(avg_series))

# Plot the results
sf.plot_series(time, series)
sf.plot_series(time[:len(avg_series)], avg_series)
plt.show()
```

```python
# See an autocorrelated series
ar = sf.autocorrelation(time, 10)
sf.plot_series(time, ar)
```
